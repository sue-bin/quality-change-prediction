```python
import pandas as pd
import numpy as np
import datetime as dt
from tqdm import tqdm
import re

def make_datetime(x):
    # string 타입의 Time column을 datetime 타입으로 변경
    x     = str(x)
    year  = int(x[:4])
    month = int(x[4:6])
    day   = int(x[6:8])
    hour  = int(x[8:10])
    minute  = int(x[10:12])
    sec  = int(x[12:])
    return dt.datetime(year, month, day, hour,minute)

def string2num(x):
    # (,)( )과 같은 불필요한 데이터 정제
    x = re.sub(r"[^0-9]+", '', str(x))
    if x =='':
        return 0
    else:
        return int(x)
    
def string2num2(x):
    # (,)( )과 같은 불필요한 데이터 정제
    x = re.sub(r"[^0-9]+", '', str(x))
    return int(x)

PATH = "Jupyter Temp/lg 공모전/data/"
```

```python
train_err  = pd.read_csv(PATH+'train_err_data.csv')
train_err_qual  = pd.read_csv(PATH+'train_quality_data.csv')
train_err_prob  = pd.read_csv(PATH+'train_problem_data.csv')
```

```python
test_err  = pd.read_csv(PATH+'test_err_data.csv')
test_qual  = pd.read_csv(PATH+'test_quality_data.csv')
```

```python
def make_datetime(x):
    # string 타입의 Time column을 datetime 타입으로 변경
    x     = str(x)
    year  = int(x[:4])
    month = int(x[4:6])
    day   = int(x[6:8])
    # hour  = int(x[8:10])
    # mim  = int(x[10:12])
    # sec  = int(x[12:])
    return dt.datetime(year, month, day)

train_err['datetime'] = train_err['time'].apply(make_datetime)
test_err['datetime'] = test_err['time'].apply(make_datetime)

weekends=["2020-11-07", "2020-11-08", "2020-11-14", "2020-11-15", "2020-11-21", "2020-11-22", "2020-11-28", "2020-11-29"]

### train
train_err_04=train_err[(train_err["datetime"]== weekends[0]) & (train_err["errtype"]==4)]
train_err_28=train_err[(train_err["datetime"]== weekends[0]) & (train_err["errtype"]==28)]
train_err_29=train_err[(train_err["datetime"]== weekends[0]) & (train_err["errtype"]==29)]
train_err_35=train_err[(train_err["datetime"]== weekends[0]) & (train_err["errtype"]==35)]
for i in range(len(weekends)):
  train_err_04_=train_err[(train_err["datetime"]== weekends[i]) & (train_err["errtype"]==4)]
  train_err_04 = pd.concat([train_err_04,train_err_04_], ignore_index=True)
  train_err_28_=train_err[(train_err["datetime"]== weekends[i]) & (train_err["errtype"]==28)]
  train_err_28 = pd.concat([train_err_28,train_err_28_], ignore_index=True)
  train_err_29_=train_err[(train_err["datetime"]== weekends[i]) & (train_err["errtype"]==29)]
  train_err_29 = pd.concat([train_err_29,train_err_29_], ignore_index=True) 
  train_err_35_=train_err[(train_err["datetime"]== weekends[i]) & (train_err["errtype"]==35)]
  train_err_35 = pd.concat([train_err_35,train_err_35_], ignore_index=True)
del train_err_04['time']
del train_err_28['time']
del train_err_29['time']
del train_err_35['time']

train_err_04 = train_err_04.groupby('user_id').sum()
train_err_04.rename(columns = {'errtype' : 'weekend_04'}, inplace = True)
train_err_28 = train_err_28.groupby('user_id').sum()
train_err_28.rename(columns = {'errtype' : 'weekend_28'}, inplace = True)
train_err_29 = train_err_29.groupby('user_id').sum()
train_err_29.rename(columns = {'errtype' : 'weekend_29'}, inplace = True)
train_err_35 = train_err_35.groupby('user_id').sum()
train_err_35.rename(columns = {'errtype' : 'weekend_35'}, inplace = True)
train_weekend = pd.concat([train_err_04, train_err_28, train_err_29, train_err_35],axis=1)
train_weekend.fillna(0,inplace=True)
print(train_weekend)

### test
test_err_04=test_err[(test_err["datetime"]== weekends[0]) & (test_err["errtype"]==4)]
test_err_28=test_err[(test_err["datetime"]== weekends[0]) & (test_err["errtype"]==28)]
test_err_29=test_err[(test_err["datetime"]== weekends[0]) & (test_err["errtype"]==29)]
test_err_35=test_err[(test_err["datetime"]== weekends[0]) & (test_err["errtype"]==35)]
for i in range(len(weekends)):
  test_err_04_=test_err[(test_err["datetime"]== weekends[i]) & (test_err["errtype"]==4)]
  test_err_04 = pd.concat([test_err_04,test_err_04_], ignore_index=True)
  test_err_28_=test_err[(test_err["datetime"]== weekends[i]) & (test_err["errtype"]==28)]
  test_err_28 = pd.concat([test_err_28,test_err_28_], ignore_index=True)
  test_err_29_=test_err[(test_err["datetime"]== weekends[i]) & (test_err["errtype"]==29)]
  test_err_29 = pd.concat([test_err_29,test_err_29_], ignore_index=True) 
  test_err_35_=test_err[(test_err["datetime"]== weekends[i]) & (test_err["errtype"]==35)]
  test_err_35 = pd.concat([test_err_35,test_err_35_], ignore_index=True)
del test_err_04['time']
del test_err_28['time']
del test_err_29['time']
del test_err_35['time']

test_err_04 = test_err_04.groupby('user_id').sum()
test_err_04.rename(columns = {'errtype' : 'weekend_04'}, inplace = True)
test_err_28 = test_err_28.groupby('user_id').sum()
test_err_28.rename(columns = {'errtype' : 'weekend_28'}, inplace = True)
test_err_29 = test_err_29.groupby('user_id').sum()
test_err_29.rename(columns = {'errtype' : 'weekend_29'}, inplace = True)
test_err_35 = test_err_35.groupby('user_id').sum()
test_err_35.rename(columns = {'errtype' : 'weekend_35'}, inplace = True)
test_weekend = pd.concat([test_err_04, test_err_28, test_err_29, test_err_35],axis=1)
test_weekend.fillna(0,inplace=True)
print(test_weekend)
```

```python
train_weekend = train_weekend.reset_index()
test_weekend = test_weekend.reset_index()
```
