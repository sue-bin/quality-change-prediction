```python
def make_datetime(x):
    # string íƒ€ìž…ì˜ Time columnì„ datetime íƒ€ìž…ìœ¼ë¡œ ë³€ê²½
    x     = str(x)
    year  = int(x[:4])
    month = int(x[4:6])
    day   = int(x[6:8])
    hour  = int(x[8:10])
    minute  = int(x[10:12])
    sec  = int(x[12:])
    return dt.datetime(year, month, day, hour,minute)
```

```python
# ë³€ìˆ˜ ì´ë¦„ ë³€ê²½
# error  -> train_x
# problem-> train_y

train_x = train.drop(columns = ["user_id", "update", "label", "err_8", "err_9"]).values
train_y = train.label.values
print(train_x.shape)
print(train_y.shape)
```

```python
# Train
#-------------------------------------------------------------------------------------
# validation auc scoreë¥¼ í™•ì¸í•˜ê¸° ìœ„í•´ ì •ì˜
def f_pr_auc(probas_pred, y_true):
    labels=y_true.get_label()
    p, r, _ = precision_recall_curve(labels, probas_pred)
    score=auc(r,p) 
    return "pr_auc", score, True
#-------------------------------------------------------------------------------------
models     = []
recalls    = []
precisions = []
auc_scores   = []
threshold = 0.5
# íŒŒë¼ë¯¸í„° ì„¤ì •
params =      {
                'boosting_type' : 'gbdt',
                'objective'     : 'binary',
                'metric'        : 'auc',
                'feature_fraction' : 0.8,
                'bagging_fraction': 0.7, 
                'bagging_freq': 5,
                'seed': 1015,
                'learning_rate' : 0.01
                }
#-------------------------------------------------------------------------------------
# 5 Kfold cross validation
k_fold = KFold(n_splits=5, shuffle=True, random_state=718917)
for train_idx, val_idx in k_fold.split(train_x):

    # split train, validation set
    X = train_x[train_idx]
    y = train_y[train_idx]
    valid_x = train_x[val_idx]
    valid_y = train_y[val_idx]

    d_train= lgb.Dataset(X, y)
    d_val  = lgb.Dataset(valid_x, valid_y)
    
    #run traning
    model = lgb.train(
                        params,
                        train_set       = d_train,
                        num_boost_round = 1500,
                        valid_sets      = d_val,
                        #feval           = f_pr_auc,
                        verbose_eval    = 100,
                        early_stopping_rounds = 300
                       )
    
    # cal valid prediction
    valid_prob = model.predict(valid_x)
    valid_pred = np.where(valid_prob > threshold, 1, 0)
    
    # cal scores
    recall    = recall_score(    valid_y, valid_pred)
    precision = precision_score( valid_y, valid_pred)
    auc_score = roc_auc_score(   valid_y, valid_prob)

    # append scores
    models.append(model)
    recalls.append(recall)
    precisions.append(precision)
    auc_scores.append(auc_score)

    print('==========================================================')
```
