```python
import pandas as pd
```

```python
train = pd.read_csv("Jupyter Temp/lg 공모전/data/0126_train_xy.csv")
test = pd.read_csv("Jupyter Temp/lg 공모전/data/0126_test_x.csv")
train_w = pd.read_csv("Jupyter Temp/lg 공모전/data/0126_train_weekend.csv")
test_w = pd.read_csv("Jupyter Temp/lg 공모전/data/0126_test_weekend.csv")
```

```python
train = train.merge(train_w,how="left",on="user_id")
test = test.merge(test_w,how="left",on="user_id")
```

```python
train.fillna(0,inplace=True)
test.fillna(0,inplace=True)
```
